```python
#define minimum/maximum threshold criteria
number_of_tweets = 500
tweet_count_min = 5
tweet_length_min = 30
```

```python
%%notify
#create raw initial dataframe
tweets_un = pd.DataFrame({})
errors = []

for a in range(0, len(df_investors.twitter_handle)):
    tweets = []
    lang = []
    id = []
    user_name = []
    un = df_investors.iloc[a,2]
    
    try:
        for b in tweepy.Cursor(api.user_timeline, id=un, tweet_mode="extended").items(number_of_tweets):
            tweets.append(b.full_text)
            lang.append(b.lang)
            id.append(b.id)

        for c in range(0,len(tweets)):
            user_name.append(df_investors.iloc[a,2])

        df1 = pd.DataFrame({'tweet': tweets})
        df2 = pd.DataFrame({'un': user_name})
        df3 = pd.DataFrame({'lang': lang})
        df4 = pd.DataFrame({'id': id})


        df1.reset_index(drop=True, inplace=True)
        df2.reset_index(drop=True, inplace=True)
        df3.reset_index(drop=True, inplace=True)
        df4.reset_index(drop=True, inplace=True)
        df5 = pd.concat([df1, df2, df3, df4], axis=1)

        tweets_un = tweets_un.append(df5)
        tweets_un = tweets_un[~tweets_un['tweet'].str.contains("RT")]
        tweets_un.reset_index(drop=True, inplace=True)
        print(a, un)
        
    except Exception:
        print('error!')
        errors.append(df_investors.iloc[a,2])
        pass
```

```python
%%notify
#create a table counting the tweet count for each user
un = tweets_un['un'].unique()
count = []


for a in range(0, len(un)):
    counter = 0
    counter = len(tweets_un[tweets_un['un'].str.contains(un[a])])

    count.append(counter)
   
df1 = pd.DataFrame({'un':un})
df2 = pd.DataFrame({'tweet_count':count})
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
tweet_counts = pd.concat([df1, df2], axis=1)
```

```python
#clean tweets df based on criteria
#any tweets that have fewer characters than tweet_length_min are excluded
tweets_un = tweets_un[tweets_un['tweet'].str.len() > tweet_length_min]

#any users that have fewer tweets than tweet_count_min are excluded
keep_tweet_count = tweet_counts[tweet_counts['tweet_count'] > tweet_count_min].un.tolist()

df_clean = pd.DataFrame({})
for i in range(0,len(keep_tweet_count)):
    holder = tweets_un[tweets_un['un'].str.contains(keep_tweet_count[i])]
    df_clean = df_clean.append(holder)
    df_clean.reset_index(drop=True, inplace=True)

#remove unknown language types --> generally just mentions
df_clean = df_clean[(df_clean['lang']!='und')]
```
